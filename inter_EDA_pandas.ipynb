mean revenues of each country.


```
avg_revenues_of_each_country={}
countries= f500['country'].unique()
for country in countries:
  select_row= f500[f500['country']==country]
  mean_revenues= select_row['revenues'].mean()
  avg_revenues_of_each_country[country]=mean_revenues


avg_revenues_of_each_country
```

extracting top employer of each country

```
top_employer_by_country={}
countries= f500['country'].unique()
for c in countries:
    selected_rows= f500[f500['country']==c]
    sorted_rows= selected_rows.sort_values('employees', ascending=False)
    top_employer= sorted_rows.iloc[0]
    employer_name= top_employer['company']
    top_employer_by_country[c]=employer_name
top_employer_by_country
```

```
dataframe= {'Country':top_employer_by_country.keys(), 
            'Company': top_employer_by_country.values()}
top_employed= pd.DataFrame(dataframe)
top_employed
```

top_roa_by_sector

```
f500['roa']= f500['profits']/f500['assets']
top_roa_by_sector={}
for sector in f500['sector']:
    select_row= f500[f500['sector']==sector]
    sorted_rows= select_row.sort_values('roa', ascending=False)
    top_roa= sorted_rows.iloc[0]
    top_roa_value_company= top_roa['company']
    top_roa_by_sector[sector]= top_roa_value_company

top_roa_by_sector
```
